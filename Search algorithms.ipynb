```python
from geopy.distance import distance
```

```python
neighbours={
    nairobi:[nakuru,thika,mombasa,malindi],
    mombasa:[malindi,nairobi,marsabit],
    kisumu:[eldoret,kakamega,kericho],
    nakuru:[kericho,nairobi],
    eldoret:[kericho,kisumu,kakamega],
    kakamega:[kisumu,eldoret],
    marsabit:[thika,mombasa,malindi],
    kericho:[kisumu,eldoret,nakuru],
    thika:[nairobi,marsabit],
    malindi:[nairobi,mombasa,marsabit]
}
```

```python
from queue import Queue
```

```python
def bfs(start, end):
    global neighbours
    # maintain a queue of paths
    queue = []
    # push the first path into the queue
    queue.append([start])
    while queue:
        # get the first path from the queue
        path = queue.pop(0)
        # get the last node from the path
        node = path[-1]
        # path found
        if node == end:
            return path
        # enumerate all adjacent nodes, construct a 
        # new path and push it into the queue
        for neighbour in neighbours.get(node, []):
            new_path = list(path)
            new_path.append(neighbour)
            queue.append(new_path)
```

```python
bfs_output=bfs(eldoret,thika)
print(bfs_output)
```

```python
bfs_length = len(bfs_output)

total_distance = 0

m1 = folium.Map(location = nairobi, width = 1000, height = 550)
folium.Marker(nairobi).add_to(m1)
folium.Marker(mombasa).add_to(m1)
folium.Marker(kisumu).add_to(m1)
folium.Marker(nakuru).add_to(m1)
folium.Marker(eldoret).add_to(m1)
folium.Marker(kakamega).add_to(m1)
folium.Marker(marsabit).add_to(m1)
folium.Marker(kericho).add_to(m1)
folium.Marker(thika).add_to(m1)
folium.Marker(malindi).add_to(m1)

for i in range(bfs_length - 1):
    total_distance += distance(bfs_output[i], bfs_output[i+1]).km
    folium.PolyLine((bfs_output[i], bfs_output[i+1])).add_to(m1)
```

```python
visited = set()  # Set to keep track of visited nodes of graph.
output_list=[]
def dfs(node, end):  # function for dfs
    global visited,neighbours
    if node not in visited:
        output_list.append(node)
        # We check if the current node is the one for which we are looking for
        if node == end:
            return True

        visited.add(node)
        for neighbour in neighbours[node]:
            # If we found the node, we break out from the loop
            if dfs(neighbour, end):
                return output_list
        # If we did not find the node we were looking for, we return False
        return False  
```

```python
tree={
    nairobi:[[nakuru,distance(nairobi,nakuru).km],[thika,distance(nairobi,thika).km],[mombasa,distance(nairobi,mombasa).km],[malindi,distance(nairobi,malindi).km]],
    mombasa:[[malindi,distance(mombasa,malindi).km],[nairobi,distance(mombasa,nairobi).km],[marsabit,distance(mombasa,marsabit).km]],
    kisumu:[[eldoret,distance(kisumu,eldoret).km],[kakamega,distance(kisumu,kakamega).km],[kericho,distance(kisumu,kericho).km]],
    nakuru:[[kericho,distance(nakuru,kericho).km],[nairobi,distance(nakuru,nairobi).km]],
    eldoret:[[kericho,distance(eldoret,kericho).km],[kisumu,distance(eldoret,kisumu).km],[kakamega,distance(eldoret,kakamega).km]],
    kakamega:[[kisumu,distance(kakamega,kisumu).km],[eldoret,distance(kakamega,eldoret).km]],
    marsabit:[[thika,distance(marsabit,thika).km],[mombasa,distance(marsabit,mombasa).km],[malindi,distance(marsabit,malindi).km]],
    kericho:[[kisumu,distance(kericho,kisumu).km],[eldoret,distance(kericho,eldoret).km],[nakuru,distance(kericho,nakuru).km]],
    thika:[[nairobi,distance(thika,nairobi).km],[marsabit,distance(thika,marsabit).km]],
    malindi:[[nairobi,distance(malindi,nairobi).km],[mombasa,distance(malindi,mombasa).km],[marsabit,distance(malindi,marsabit).km]]
}
```

```python
heuristic={
    nairobi:distance(nairobi,malindi).km,
    mombasa:distance(mombasa,malindi).km,
    kisumu:distance(kisumu,malindi).km,
    nakuru:distance(nakuru,malindi).km,
    eldoret:distance(eldoret,malindi).km,
    kakamega:distance(kakamega,malindi).km,
    marsabit:distance(marsabit,malindi).km,
    kericho:distance(kericho,malindi).km,
    thika:distance(thika,malindi).km,
    malindi: 0
}
```

```python
def AStar_search(start,finish):
    cost={start:0}
    global tree,heuristic,neighbours
    closed = []             # closed nodes
    opened = [[start, distance(start,finish).km]]     # opened nodes

    '''find the visited nodes'''
    while True:
        fn = [i[1] for i in opened]     # fn = f(n) = g(n) + h(n)
        chosen_index = fn.index(min(fn))
        node = opened[chosen_index][0]  # current node
        closed.append(opened[chosen_index])
        del opened[chosen_index]
        if closed[-1][0] == finish:        # break the loop if node G has been found
            break
        for item in tree[node]:
            if item[0] in [closed_item[0] for closed_item in closed]:
                continue
            cost.update({item[0]: cost[node] + item[1]})            # add nodes to cost dictionary
            fn_node = cost[node] + heuristic[item[0]] + item[1]     # calculate f(n) of current node
            temp = [item[0], fn_node]
            opened.append(temp)                                     # store f(n) of current node in array opened

    '''find optimal sequence'''
    trace_node = finish                       # correct optimal tracing node, initialize as node G
    optimal_nodes = []
    optimal_nodes.append([start])
    optimal_sequence = [finish]                # optimal node sequence
    for i in range(len(closed)-2, -1, -1):
        check_node = closed[i][0]           # current node
        if trace_node in [children[0] for children in tree[check_node]]:
            children_costs = [temp[1] for temp in tree[check_node]]
            children_nodes = [temp[0] for temp in tree[check_node]]

            '''check whether h(s) + g(s) = f(s). If so, append current node to optimal sequence
            change the correct optimal tracing node to current node'''
            if cost[check_node] + children_costs[children_nodes.index(trace_node)] == cost[trace_node]:
                optimal_sequence.append(check_node)
                trace_node = check_node
    while optimal_nodes:
        path = optimal_nodes.pop(0)
        node = path[-1]
        if node == finish:
            return path
        for adjacent in neighbours.get(node, []):
            optimal_path = list(path)
            optimal_path.append(adjacent)
            optimal_nodes.append(optimal_path)              
    return optimal_nodes
```

```python
optimal_nodes = AStar_search(eldoret,thika)
```

```python
a_star_length = len(optimal_nodes)

total_distance = 0

m3 = folium.Map(location = nairobi, width = 1000, height = 550)
folium.Marker(nairobi).add_to(m3)
folium.Marker(mombasa).add_to(m3)
folium.Marker(kisumu).add_to(m3)
folium.Marker(nakuru).add_to(m3)
folium.Marker(eldoret).add_to(m3)
folium.Marker(kakamega).add_to(m3)
folium.Marker(marsabit).add_to(m3)
folium.Marker(kericho).add_to(m3)
folium.Marker(thika).add_to(m3)
folium.Marker(malindi).add_to(m3)

for i in range(a_star_length - 1):
    total_distance += distance(optimal_nodes[i], optimal_nodes[i+1]).km
    folium.PolyLine((optimal_nodes[i], optimal_nodes[i+1])).add_to(m3)
```
